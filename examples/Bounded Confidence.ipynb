```python
import numpy as np
import matplotlib.pyplot as plt
import OpinionModel
rng = np.random.default_rng()
```

##### $\phi=0.05 \cdot I_{0.5}$

```python
model = OpinionModel.Deterministic()
model.phi = lambda h : 0.01 * (h<=0.5)
steps, nodes = model.generate(steps=10000)
for i in range(len(nodes[:,0])):
    plt.plot(steps, nodes[i,:])
plt.title("$\phi=0.01 \cdot I_{0.5}$")
plt.xscale('log')
plt.show()

model.phi = lambda h : 0.01 * (h<=0.5)
x = np.array([1,0.95, 0.9, 0.85, 0.8, 0.2, 0.15, 0.1, 0.05, 0])
steps, nodes = model.generate(x=x, steps=10000)
for i in range(len(nodes[:,0])):
    plt.plot(steps, nodes[i,:])
plt.title("$\phi=0.01 \cdot I_{0.5}$")
plt.xscale('log')
plt.show()

model.phi = lambda h : 0.01 * (h<=0.5)
x = rng.random(size=10)
steps, nodes = model.generate(x=x, steps=10000)
for i in range(len(nodes[:,0])):
    plt.plot(steps, nodes[i,:])
plt.title("$\phi=0.01 \cdot I_{0.5}$")
plt.xscale('log')
```
