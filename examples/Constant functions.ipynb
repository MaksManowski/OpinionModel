```python
import numpy as np
import matplotlib.pyplot as plt
import OpinionModel
rng = np.random.default_rng()
```

##### $\phi=0.05$

```python
model = OpinionModel.Deterministic()
model.phi = lambda h : 0.05 + 0 * h
steps, nodes = model.generate(steps=10000)
for i in range(len(nodes[:,0])):
    plt.plot(steps, nodes[i,:])
plt.title("$\phi=0.05$")
plt.xscale('log')
```

Randomised x

```python
model = OpinionModel.Deterministic()
model.phi = lambda h : 0.001 + 0 * h
x = rng.random(size=25)/1.6
x = np.append(x, 1)
steps, nodes = model.generate(x=x, steps=10000)
for i in range(len(nodes[:,0])):
    plt.plot(steps, nodes[i,:])
plt.title("$\phi=0.001$")
plt.xscale('log')
```
